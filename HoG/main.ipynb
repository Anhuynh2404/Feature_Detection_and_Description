```python
from PIL import Image
import HOG
import matplotlib.pyplot as plt
import numpy as np
```

```python
image_path = "/home/an/an_workplace/Lab_CV/Data/ozil.png"
try:
    Image.open(image_path)
except FileNotFoundError:
    print(f"File not found: {image_path}")

img_loaded = HOG.load_and_preprocess_image(image_path)
```

```python
horizontal_mask = np.array([-1, 0, 1])
vertical_mask = np.array([[-1],
                             [0],
                             [1]])
```

```python
gradient_y, gradient_x = np.gradient(img_loaded)
horizontal_gradient = HOG.calculate_gradient(img_loaded, horizontal_mask)
vertical_gradient = HOG.calculate_gradient(img_loaded, vertical_mask)
```

```python


grad_magnitude = HOG.gradient_magnitude(horizontal_gradient, vertical_gradient)
grad_direction = HOG.gradient_direction(horizontal_gradient, vertical_gradient)
titles = ['X-gradient', 'Y-gradient', 'Gradient Magnitude']
images = [horizontal_gradient, vertical_gradient, grad_magnitude]

plt.figure(figsize=(12, 4))
for i in range(3):
    plt.subplot(1, 3, i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')
plt.tight_layout()
plt.show()
```
